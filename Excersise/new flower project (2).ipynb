```python
from sklearn import datasets
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import seaborn as sns
from sklearn import svm 
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split,learning_curve
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
#importing data
iris=datasets.load_iris()
iris
```

```python
#creating data frame from the iris data set
df=pd.DataFrame(iris.data,columns=["sepal length","sepal width","petal length","petal width"])
df
```

```python
df2=pd.DataFrame(iris.target, columns=["Class"])
df2
#el target dah hwwa tasnif el warda de no3ha eh
```

```python
dffinal=pd.concat([df,df2],axis=1)
'''lamma makontesh mesammi 2sma2 el columns sa3et-ha kan 3ndi two columns lihom nafs el esm w kan esmo (0),
2bl el concate nesammi el columns beta3tna w b3dan n3mel concat'''
dffinal
```

```python
# plot correlation matrix to show relations between features
sns.heatmap(dffinal.corr(),annot=True)
```

```python
#we should define the sets of train
X = dffinal[["sepal length","sepal width","petal length","petal width"]] # X is the feature
y = dffinal['Class'] # y is the label
```

```python
# train test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
# test_size=0.25 , means that 25% of the data is for test and the rest is for training
# train_size=0.75 , means that 75% of the data is for training and the rest is for testing
# random_state=101 , when random state is set by a number the same random data is taken every time it is run
```

```python
model1=LogisticRegression()
model1.fit(X_train,y_train)
```

```python
y_pred = model1.predict(X_test)
```

```python
#getting the accuracy
accuracy_score(y_test,y_pred)
```

```python
#classification report
print(classification_report(y_test, y_pred))
```

```python
#confusion matrix
con_mat=confusion_matrix(y_test, y_pred)
con_mat
```

```python
sns.heatmap(con_mat,annot=True)
# 3shan nersem el confusion matrex bel heatmap , lazem no7ttaha fe dataframe
```

```python
dffinal.drop(['petal length'],axis=1,inplace=True)
accuracy_score(y_test,y_pred)
```

```python
model2=SVC(kernel='rbf' , gamma=100)
#el gamma kol ma tezid el model yb2a mesh masmo7lo y8lat , y3ni bydya2 3la el points
#the gamma range is from 0.1:100
#gamma is used with rbf
model2.fit(X_train,y_train)
```

```python
#for visulaizing the kernal of the SVM

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from sklearn import neighbors, datasets

n_neighbors = 15

# import some data to play with
iris = datasets.load_iris()

# we only take the first two features. We could avoid this ugly
# slicing by using a two-dim dataset
X = iris.data[:, :2]
y = iris.target

h = .02  # step size in the mesh

# Create color maps
cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])

# we create an instance of Neighbours Classifier and fit the data.
clf = model2
clf.fit(X, y)

# Plot the decision boundary. For that, we will assign a color to each
# point in the mesh [x_min, x_max]x[y_min, y_max].
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                     np.arange(y_min, y_max, h))
Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])

# Put the result into a color plot
Z = Z.reshape(xx.shape)
plt.figure()
plt.pcolormesh(xx, yy, Z, cmap=cmap_light)

# Plot also the training points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold,
            edgecolor='k', s=20)
plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.title("test")

plt.show()
```
